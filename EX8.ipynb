```python
import numpy as np 
import matplotlib.pyplot as plt

values= np.random.normal(500,35, 50000)
print(values.shape)
plt.hist(values,500,density=True)
plt.show()
```

```python
s = np.random.uniform(-10,50,3000)
count, bins, ignored = plt.hist(s, 70,density=True)
plt.show()
```

```python
pois = np.random.poisson(9.4, 50000)
plt.hist(pois, 10,density=True)
plt.show()
```

```python
n=500
p=0.0078
binom = np.random.binomial(n, p, 10000)
plt.hist(binom,20,density=True)
plt.show()
```

```python
target = 650
beta = 1.0/target
Y = np.random.exponential(beta, 10000)
plt.hist(Y, bins=200,lw=0,alpha=.8)
plt.plot([0,max(Y)],[target,target],'r--')
plt.ylim(0,target*1.1)
plt.show()
```
